```python
import cv2
import matplotlib.pyplot as plt

# Specify the path to the image file
image_path = r"C:\Users\ramco\OneDrive\Desktop\Mtech Docs\Second Sem\Image and Video\PG_lab\Photo1.jpg"

# Load the image in BGR format (OpenCV default)
image_bgr = cv2.imread(image_path)

# Check if the image was loaded successfully
if image_bgr is None:
    raise ValueError("Image not loaded. Check the file path and format.")

# Convert the BGR image to RGB format
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Define the text to add
text = "I am proud to be an Indian"

# Define text properties
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 7
font_color = (0, 0, 255)  # Red color in RGB
thickness = 10
line_type = cv2.LINE_AA

# Get text size to position the text properly
(text_width, text_height), baseline = cv2.getTextSize(text, font, font_scale, thickness)
height, width, _ = image_rgb.shape

# Calculate the position for centered text
text_x = (width - text_width) // 2
text_y = (height + text_height) // 2

# Add text to the image
cv2.putText(image_rgb, text, (text_x, text_y), font, font_scale, font_color, thickness, line_type)

# Save the modified RGB image to a new file
saved_image_path = r"C:\Users\ramco\OneDrive\Desktop\Mtech Docs\Second Sem\Image and Video\PG_lab\Photo1 - Copy.jpg"
cv2.imwrite(saved_image_path, cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR))  # Convert back to BGR for saving

# Display the modified RGB image with text
plt.figure(figsize=(7.5, 5))
plt.imshow(image_rgb)
plt.title('Modified RGB Image with Centered Text')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Specify the path to the image file
image_path = r"C:\Users\ramco\OneDrive\Desktop\Mtech Docs\Second Sem\Image and Video\PG_lab\Photo1.jpg"

# Load the image in BGR format (OpenCV default)
image_bgr = cv2.imread(image_path)

# Check if the image was loaded successfully
if image_bgr is None:
    raise ValueError("Image not loaded. Check the file path and format.")

# Convert the BGR image to RGB format
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Define the text to add
text = "I am proud to be an Indian"

# Define text properties
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 3.5
font_color = (0, 0, 255)  # White color in RGB
thickness = 3
line_type = cv2.LINE_AA

# Get text size to position the text properly
(text_width, text_height), baseline = cv2.getTextSize(text, font, font_scale, thickness)
height, width, _ = image_rgb.shape

# Define circle properties
center_x, center_y = width // 2, height // 2
radius = 800

# Draw the circle on the image
cv2.circle(image_rgb, (center_x, center_y), radius, (0, 255, 0), 7)  # Green circle with thickness of 5

# Calculate the position for centered text inside the circle
text_x = center_x - text_width // 2
text_y = center_y + text_height // 2

# Add text inside the circle
cv2.putText(image_rgb, text, (text_x, text_y), font, font_scale, font_color, thickness, line_type)

# Save the modified RGB image to a new file
saved_image_path = r"C:\Users\ramco\OneDrive\Desktop\Mtech Docs\Second Sem\Image and Video\PG_lab\Photo1 - Copy with Circle.jpg"
cv2.imwrite(saved_image_path, cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR))  # Convert back to BGR for saving

# Display the modified RGB image with circle and text
plt.figure(figsize=(7.5, 5))
plt.imshow(image_rgb)
plt.title('Modified RGB Image with Circle and Text')
plt.axis('off')
plt.show()
```
